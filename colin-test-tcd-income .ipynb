```python
obj_train = train.select_dtypes(include=['object']).copy()
obj_train.head()


obj_test = test.select_dtypes(include=['object']).copy()
obj_test.head()
```

```python
temp_train_df = pd.get_dummies(obj_train, columns=["Profession", "Hair Color", "Country"])
encoded_train_df = temp_train_df.select_dtypes(include=['uint8']).copy()
encoded_train_df.head()

temp_test_df = pd.get_dummies(obj_test, columns=["Profession", "Hair Color", "Country"])
encoded_test_df = temp_test_df.select_dtypes(include=['uint8']).copy()
encoded_test_df.head()
```

```python
train = train.drop(['Profession', 'Hair Color', 'Country'], axis = 1)
complete_train_data = pd.concat([train, encoded_train_df], axis =1)
complete_train_data.head()

test = test.drop(['Profession', 'Hair Color', 'Country'], axis = 1)
complete_test_data = pd.concat([test, encoded_test_df], axis =1)
complete_test_data.head()
```

```python
X = complete_train_data.drop(['Income in EUR', 'Instance'], axis =1)
#print(X.head())
Y = complete_train_data['Income in EUR']
#Y.shape, X.shape
#X.info(), Y.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size = 0.2, random_state=3)
#y_train.head()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(complete_test_data.shape)
```

```python
from sklearn import linear_model
lin_model = linear_model.LinearRegression()
lin_model.fit(x_train, y_train)
```

```python

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
```

```python
y_train_predict = lin_model.predict(x_train)

rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
# model evaluation for testing set
#predictions = my_model.predict(test_X)
y_test_predict = lin_model.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))

```
